```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load your two datasets into Pandas DataFrames

transaction_data = pd.read_csv(r'C:\Users\Olamilekan .A. David\OneDrive\Documents\Data Analytics\Quantium\Customer Analysis\QVI_transaction_data_excel_cleaned.csv')
purchase_behavior = pd.read_csv(r'C:\Users\Olamilekan .A. David\OneDrive\Documents\Data Analytics\Quantium\Customer Analysis\QVI_purchase_behaviour_excel_cleaned.csv')
```

```python
import re
from collections import Counter

# Extract words from 'PROD_NAME' column and remove digits and special characters
words = chips_data['PROD_NAME'].str.lower().str.replace(r'[^a-z\s]', '', regex=True).str.split()

# Combine all the words into a single list
all_words = [word for sublist in words for word in sublist]

# Create a Counter to count the frequency of each word
word_counts = Counter(all_words)

# Sort the distinct words by frequency of occurrence
sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Print the sorted words and their frequencies
for word, frequency in sorted_words:
    print(f"{word}: {frequency}")
```

```python
import scipy.stats as stats

# Filter the data to include only midage and young singles and couples
filtered_data = chips_data[
    (chips_data['LIFESTAGE'].isin(['MIDAGE SINGLES/COUPLES', 'YOUNG SINGLES/COUPLES'])) & 
    (chips_data['PREMIUM_CUSTOMER'].isin(['Mainstream', 'Budget', 'Premium']))
]

# Group the data by LIFESTAGE and PREMIUM_CUSTOMER and calculate the average price per unit
grouped_data = filtered_data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['TOT_SALES'].sum() / filtered_data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['PROD_QTY'].sum()

# Extract the average price per unit for Mainstream and Premium groups
mainstream_avg_price = grouped_data['MIDAGE SINGLES/COUPLES']['Mainstream']
premium_avg_price = grouped_data['MIDAGE SINGLES/COUPLES']['Premium']

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(
    filtered_data[filtered_data['PREMIUM_CUSTOMER'] == 'Mainstream']['TOT_SALES'] / filtered_data[filtered_data['PREMIUM_CUSTOMER'] == 'Mainstream']['PROD_QTY'],
    filtered_data[filtered_data['PREMIUM_CUSTOMER'] == 'Premium']['TOT_SALES'] / filtered_data[filtered_data['PREMIUM_CUSTOMER'] == 'Premium']['PROD_QTY'],
    equal_var=False  # Assuming unequal variances
)

# Check if the difference is statistically significant
alpha = 0.05  # Set the significance level
if p_value < alpha:
    print("The difference in average price per unit is statistically significant.")
else:
    print("The difference in average price per unit is not statistically significant.")

# Print the t-statistic and p-value
print(f"t-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
# Filter the data to include only Mainstream young singles/couples
mainstream_young = chips_data[
    (chips_data['LIFESTAGE'] == 'YOUNG SINGLES/COUPLES') &
    (chips_data['PREMIUM_CUSTOMER'] == 'Mainstream')
]

# Calculate the total quantity of chips purchased for each pack size within the Mainstream young singles/couples segment
pack_size_preference_mainstream = mainstream_young.groupby('PROD_QTY')['TXN_ID'].count()

# Calculate the total quantity of chips purchased for each pack size for the entire population
pack_size_preference_total = chips_data.groupby('PROD_QTY')['TXN_ID'].count()

# Combine both results into a DataFrame for easier comparison
pack_size_preference_df = pd.DataFrame({
    'Mainstream_Young_Singles_Couples': pack_size_preference_mainstream,
    'Total_Population': pack_size_preference_total
})

# Calculate the proportion of purchases for each pack size within each group
pack_size_preference_df['Mainstream_Young_Singles_Couples_Pct'] = (pack_size_preference_df['Mainstream_Young_Singles_Couples'] /
                                                                  pack_size_preference_df['Mainstream_Young_Singles_Couples'].sum())

pack_size_preference_df['Total_Population_Pct'] = (pack_size_preference_df['Total_Population'] /
                                                    pack_size_preference_df['Total_Population'].sum())

# Print the pack size preference comparison
print("Pack Size Preference Comparison:")
print(pack_size_preference_df)
```
